# Limpando um dataset

### conhecimentos trabalhados:

* Importação de bibliotecas
* Leitura e visualização do CSV e suas informações
* Filtragem das informações
* Limpeza e ajuste das informações (Renomeação de colunas, atribuição de valores para celulas nulas)

```python
# Importando biblioteca Pandas
import pandas as pd
```

```python
# Lendo CSV e guardando em uma variavel para operações com os dados
base = pd.read_csv("/depression_disorder_data.csv")
```

```python
# Exbibe um tipo de dado da variavel, no caso a variavel base é um dataframe
type(base)
```

```python
# Visualizando a base
# Exibe todos os registros
base
```

```python
# Verificando informações da base
# Apresenta informacoes sobre o documento csv, colunas, linhas, tipos de dados inseridos em cada coluna
base.info()
```

```python
# Fucao head(), exibe os 5 primeiros registros, mas a quantidade pode variar dependendo do valor passado como parametro
base.head()
```

```python
# Verificando valores nulos
# Funcao isnull(), verifica registros em colunas estao com valores nulos. Funcao sum(), soma todas as celulas com valores nulos naquela coluna
base.isnull().sum()
```

```python
# Filtrando os registros
base[base.Gender == "Male"]
```

```python
base.City
```

```python
# renomenando colunas, esse csv não tem os nomes das colunas em formato correto, então é necessário renomear algumas colunas com a Funcao rename()
base = base.rename(columns={"Sleep Duration" : "Sleep_Duration"})
```

```python
#Filtrando base
base[base.Sleep_Duration == "Less than 5 hours"]
```

```python
base = base.rename(columns={"Financial Stress" : "financial_stress"})
```

```python
# Filtrando registros com valores nulos em financial_stress
base[base.financial_stress.isnull()]
```

```python
# Funcao loc(), Filtra pela os registros e mostra somente a coluna onde a as celulas ce encontram com valor nulo
base.loc[base.financial_stress.isnull(), "financial_stress"]
```

```python
# Atribui as celulas localizadas o valor de 0
base.loc[base.financial_stress.isnull(), "financial_stress"] = 0
```

```python
# Agora nossa base não possui valores nulos
base.isnull().sum()
```

```python
# Excluimos registros que possuem valores nulos, axis=0 pois é o eixo x, na horizontal
base = base.dropna(axis=0)
```

```python
# Deleta uma coluna, axis=1 pois é o eixo y, na vertical
base = base.drop("Job Satisfaction", axis=1)
```

```python
base = base.drop("Work Pressure", axis=1)
```

```python
# a base de dados não possui mais as colunas "Job Satisfaction" e "Work Pressure"
base.info()
```
